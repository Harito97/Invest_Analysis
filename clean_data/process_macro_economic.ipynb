```python
import pandas as pd
import matplotlib.pyplot as plt

# Giả sử df là DataFrame của bạn
df = pd.DataFrame(data)

# Tính GDP tăng trưởng
GDP_base = df['GDP_bình_quân_(VND)'][0]  # Lấy giá trị GDP của năm cơ sở (1995)
df['GDP_growth'] = (df['GDP_bình_quân_(VND)'] - GDP_base) / GDP_base  # Tính tỷ lệ tăng trưởng GDP từ năm cơ sở

# Tính CPI tích lũy
cpi_base = 1  # CPI của năm cơ sở giả sử là 1
df['CPI_cumulative'] = cpi_base  # Bắt đầu từ giá trị CPI cơ sở là 1

# Tính tích lũy CPI qua các năm
for i in range(1, len(df)):
    df['CPI_cumulative'][i] = df['CPI_cumulative'][i-1] * (1 + df['CPI'][i]/100)
df['CPI_cumulative_growth'] = (df['CPI_cumulative'] - cpi_base) / cpi_base  # Tính tỷ lệ tăng trưởng CPI từ năm cơ sở

# Vẽ biểu đồ nhóm 2: Kinh tế
fig, ax = plt.subplots(figsize=(10, 6))

# Vẽ GDP và CPI tích lũy
ax.plot(df['Year'], df['GDP_growth'] / 100, label='Tăng trưởng GDP bình quân / 100 (%)', color='b')
ax.plot(df['Year'], df['CPI_cumulative_growth'], label='CPI tích lũy (so với năm cơ sở)', color='r')

# Thêm tiêu đề và nhãn
ax.set_title('So sánh Tăng trưởng GDP Bình quân và CPI Tích lũy')
ax.set_xlabel('Năm')
ax.set_ylabel('Tỷ lệ Tăng trưởng (%) / CPI Tích lũy')

# Thêm đường lưới và hiển thị chú thích
ax.grid(False)
ax.legend()

# Hiển thị đồ thị
plt.tight_layout()
plt.savefig('../figs/macro_analysis_gdp_avg_cpi_1.png')
plt.show()
```
